## Count Words Frequency
- Write a Python function that takes a sentence as input and returns a dictionary with the frequency of each word.
- Sort the words in descending order of their frequency.

```python
def countWordsFreq(text):
    text = text.strip().lower()
    words = text.split()

    wordFreq = {}
    for word in words:
        word = "".join(ch for ch in word if ch.isalnum())
        if not word:
            continue

        if word not in wordFreq:
            wordFreq[word] = 0
        wordFreq[word] += 1

    wordFreq = dict(sorted(wordFreq.items(), key=lambda item: item[1], reverse=True))

    for key, value in wordFreq.items():
        print(f"'{key}' appeared {value} times")
```

```python
countWordsFreq(
    "The Royal Bengal tiger is a powerful and iconic apex predator, but it possesses some surprising traits. Each Bengal tiger has a unique stripe pattern, much like a human's fingerprints, which helps scientists identify individuals in the wild. These stripes also serve as excellent camouflage, making them difficult for prey to spot in the dappled light of forests and tall grasslands. Unlike many other cats, Bengal tigers are exceptional swimmers and are known to inhabit the Sundarbans mangrove forest in India and Bangladesh, the only mangrove habitat in the world where tigers are found."
)
```
