# Replace Vowels
- Replace every vowel in the input string with a hash sign (#)

```python
def replaceVowels(text):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    for char in text:
        if char in vowels:
            text = text.replace(char, "#")
    return text
```

```python
metro = "The Kolkata Metro is India's oldest metro system, opening in 1984, and is the only one in the country controlled by the Indian Railways. It features India's first underground section and is planned to have the country's first underwater metro line between Howrah and Mahakaran stations. The network is a mix of underground, elevated, and at-grade stations, using both broad-gauge and standard-gauge tracks."
replaceVowels(metro)
```
