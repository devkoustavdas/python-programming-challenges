## Check a number is Armstrong or not
An Armstrong number is a number that is equal to the sum of its own digits, each raised to the power of the number of digits.

```python
def checkArmstrongNumber(number):
    if not isinstance(number, int):
        return False
    if number<0:
        return False
    if number==0:
        return True
    num = number
    patternresult = 0
    while(num):
        reminder = num%10
        patternresult += reminder**3
        num = num // 10
    if patternresult == number:
        return True
    else:
        return False
```

```python
checkArmstrongNumber(0)
```

```python
checkArmstrongNumber(1)
```

```python
checkArmstrongNumber(10)
```

```python
checkArmstrongNumber(-10)
```

```python
checkArmstrongNumber(370)
```
