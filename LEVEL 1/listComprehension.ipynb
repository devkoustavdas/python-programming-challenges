## List comprehension

- Use list comprehension to create a new list from `[6, 2, 14, 5, 8, 30, 15, 12, 19, 24, 3, 18, 25]`
- (x^2-2x) where x is a multiple of 6

```python
a = [6, 2, 14, 5, 8, 30, 15, 12, 19, 24, 3, 18, 25]
```

```python
b = [(num**2 - 2*num) for num in a if num%2 == 0 and num%3 == 0]
```
