## Calculate LCM & HCF

- For two givcen numbers find their HCF and LCM where-
- HCF (Highest Common Factor) is the largest number that divides two or more numbers without a remainder
- LCM (Least Common Multiple) is the smallest number that is a multiple of those numbers. 

```python
def HCFandLCM(a, b):
    hcf = 0
    for i in range(1, a+1):
        if a%i == 0 and b%i == 0:
            hcf = i
    lcm = (a*b)//hcf
    return hcf, lcm
```

```python
HCF, LCM = HCFandLCM(12, 16)
print(f"HCF: {HCF}\nLCM: {LCM}")
```
