## Anagram Checker
- An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, using all the original letters exactly once.
- Take two strings and check if they are anagrams using frequency dictionaries.

```python
def anagramChecker(text1, text2):

  text1 = text1.strip().replace(" ", "").lower()
  text2 = text2.strip().replace(" ", "").lower()
  l1 = len(text1)
  l2 = len(text2)

  if (l1 != l2):
    print("Not an anagram")
    return
  else:
    l = l1
    freqText1 = {}
    freqText2 = {}

    for i in range(l):
      if text1[i] not in freqText1:
        freqText1[text1[i]] = 0
      freqText1[text1[i]] += 1

      if text2[i] not in freqText2:
        freqText2[text2[i]] = 0
      freqText2[text2[i]] += 1

    for key in freqText1:
      if freqText1[key] != freqText2[key]:
        print("Not an anagram")
        return
    print("Anagram")
    return
```

```python
anagramChecker("A Gentleman", "Elegant Man")
```

```python
anagramChecker("The Morse Code", "Here come dots")
```

```python
anagramChecker("Science", "Apple")
```

```python
anagramChecker("Race", "Care")
```
