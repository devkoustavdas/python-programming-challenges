## Expenses Split Calculator
- Input bill amounts and participants, split fairly with decimals

```python
def Expenses():
    persons_count = int(input("No. of persons in the split: "))
    persons = []
    totalExpense = 0
    for i in range(persons_count):
        name = input(f"Person {i+1} Name: ").strip()
        expense = int(input(f"Expense made by {name}: "))
        persons.append({
            "Name": name,
            "Expense": expense
        })
        print(f"{i+1}. {name} spent ₹{expense}.")
        totalExpense += expense
    print(f"\nExpenses has been noted!\nTotal Expenditure: ₹{totalExpense}\nAcross {persons_count} persons.\n")
    splitExpense(persons_count, persons, totalExpense)

def splitExpense(persons_count, persons, totalExpense):
    pay = round(totalExpense / persons_count, 1)
    print(f"SPLITS\n\nEach person must pay ₹{pay}\n")
    getMoney, giveMoney = [], []
    for i in range(persons_count):
        change = persons[i]["Expense"] - pay
        if change > 0:
            print(f"{i+1}. {persons[i]["Name"]} should get a total of ₹{round(change,1)}.")
            getMoney.append({"Name": persons[i]["Name"], "Change": round(change,1)})
        elif change < 0:
            print(f"{i+1}. {persons[i]["Name"]} should give a total of ₹{-round(change,1)}.")
            giveMoney.append({"Name": persons[i]["Name"], "Change": -round(change,1)})
        else:
            print(f"{i+1}. {persons[i]["Name"]} is settled.")

    print("\nFINAL SETTLEMENTS\n")

    getMoney.sort(key=lambda x: -x["Change"])
    giveMoney.sort(key=lambda x: -x["Change"])

    for i in range(len(giveMoney)):
        while giveMoney[i]["Change"] > 0 and len(getMoney) > 0:

            if giveMoney[i]["Change"] > getMoney[0]["Change"]:
                print(f"{giveMoney[i]['Name']} gives ₹{getMoney[0]['Change']} to {getMoney[0]['Name']}")
                giveMoney[i]["Change"] = round(giveMoney[i]["Change"] - getMoney[0]["Change"], 2)
                getMoney.pop(0)

            else:
                print(f"{giveMoney[i]['Name']} gives ₹{giveMoney[i]['Change']} to {getMoney[0]['Name']}")
                getMoney[0]["Change"] = round(getMoney[0]["Change"] - giveMoney[i]["Change"], 2)
                giveMoney[i]["Change"] = 0

                if getMoney[0]["Change"] == 0:
                    getMoney.pop(0)
```

```python
Expenses()
```
