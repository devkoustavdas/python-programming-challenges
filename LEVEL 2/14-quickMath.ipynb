```python
import random

def quickMath():
    choice = random.randint(1, 4)

    if choice == 1:
        print("Multiplication of two 2-digit numbers")
        a = random.randint(10, 99)
        b = random.randint(10, 99)
        result = a * b
        answer = int(input(f"{a} x {b} = "))
        if result == answer:
            print("Correct Answer")
        else:
            print("Wrong Answer")

    elif choice == 2:
        print("Addition of two or more numbers")
        count = random.randint(3, 4)

        if count == 3:
            a = random.randint(10, 99)
            b = random.randint(10, 199)
            c = random.randint(10, 99)
            result = a + b + c
            answer = int(input(f"{a} + {b} + {c} = "))
        else:
            a = random.randint(10, 199)
            b = random.randint(10, 199)
            c = random.randint(10, 99)
            d = random.randint(10, 99)
            result = a + b + c + d
            answer = int(input(f"{a} + {b} + {c} + {d} = "))

        if result == answer:
            print("Correct Answer")
        else:
            print("Wrong Answer")

    elif choice == 3:
        print("Multiplication of two numbers divisible by 5")
        a = random.randint(1, 50) * 5
        b = random.randint(1, 50) * 5
        result = a * b
        answer = int(input(f"{a} x {b} = "))
        if result == answer:
            print("Correct Answer")
        else:
            print("Wrong Answer")

    elif choice == 4:
        print("Percentage of a number")
        a = random.randint(1, 9) * 10
        b = random.randint(10, 100) * 10
        result = a / 100 * b
        answer = float(input(f"{a}% of {b} = "))
        if result == answer:
            print("Correct Answer")
        else:
            print("Wrong Answer")
```

```python
quickMath()
```
