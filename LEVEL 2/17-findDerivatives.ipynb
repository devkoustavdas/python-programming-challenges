## Find 1st and 2nd Derivative of a function and their Plots

Use the damped harmonic response function-
$$\boxed{
f(x) = Ae^{-\alpha x}\frac{\sin(\beta x)}{x} + C\frac{\ln(1+x^2)}{x}
}$$

Constants: $A = 2.5, \alpha = 0.7, \beta = 3.0, C = 0.8$

1. **4th-order central difference for 1st Derivative (f'(x))**

$$
\boxed{
f'(x) \approx
\frac{-f(x+2h) + 8f(x+h) - 8f(x-h) + f(x-2h)}{12h}
}$$
    
* **Order of accuracy:** $(O(h^4))$
2. **4th-order central difference for 2nd Derivative (f''(x))**
$$\boxed{
f''(x) \approx
\frac{-f(x+2h)+16f(x+h)-30f(x)+16f(x-h)-f(x-2h)}{12h^2}
}$$

* **Order of accuracy:** $(O(h^4))$

3. Finally, show graphs of the function, its single and double derivative.

```python
import math

def f(x):
    A = 2.5
    alpha = 0.7 
    beta = 3.0
    C = 0.8

    if x == 0:
        x = 10e-16

    return (A*(math.exp(-1 * alpha * x))*(math.sin(beta * x))/x + C*(math.log(1 + x**2, math.e))/x)
```

```python
sigma = 2.22 * 10e-16
```

```python
def df(x):
    h = sigma**0.5 * max(1, abs(x))
    f1 = f(x + (2 * h))
    f2 = f(x + h)
    f3 = f(x - h)
    f4 = f(x - (2 * h))
    return round((-1*f1 + 8*f2 -8*f3 + f4)/(12*h), 4)
```

```python
def ddf(x):
    h = sigma**0.25 * max(1, abs(x))
    f1 = f(x + (2 * h))
    f2 = f(x + h)
    f3 = f(x - h)
    f4 = f(x - (2 * h))
    f5 = f(x)
    return round((-1 * f1 + 16 * f2 - 30 * f5 + 16 * f3 - f4) / (12 * h**2), 4)
```

```python
df(0.2), ddf(0.2)
```

```python
df(2), ddf(2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 10000)
y = np.array([f(xi) for xi in x])

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x)")
plt.axvline(0, linestyle="--", alpha=0.4)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xx = np.linspace(-5, 5, 10000)
yy = np.array([df(xi) for xi in x])

plt.figure(figsize=(8, 5))
plt.plot(xx, yy, label="df(x)")
plt.axvline(0, linestyle="--", alpha=0.4)
plt.xlabel("x")
plt.ylabel("df(x)")
plt.title("Plot of df(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xxx = np.linspace(-5, 5, 10000)
yyy = np.array([ddf(xi) for xi in x])

plt.figure(figsize=(8, 5))
plt.plot(xxx, yyy, label="ddf(x)")
plt.axvline(0, linestyle="--", alpha=0.4)
plt.xlabel("x")
plt.ylabel("ddf(x)")
plt.title("Plot of ddf(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
