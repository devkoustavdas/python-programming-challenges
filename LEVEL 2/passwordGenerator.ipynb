**Write a function that generates a random password of a given length, ensuring it contains at least one digit, one uppercase letter, one lowercase letter, and one special character, with proper validation for input type and length.**

```python
import random

def password(l):

    if l <= 0:
        return "Invalid length"
    if l > 50:
        return "Length too long"
    if type(l) != int:
        return "Invalid type"

    password = []
    for i in range(l):
        if i == 0:
            password.append(f"{random.randint(0,9)}")
        elif i == 1:
            password.append(chr(random.randint(65, 90)))
        elif i == 2:
            password.append(chr(random.randint(97, 122)))
        elif i == 3:
            password.append(random.choice(["%", "@", "*", "#", "&", "?"]))
        else:
            choice = random.randint(1,4)
            if choice == 1:
                password.append(f"{random.randint(0,9)}")
            elif choice == 2:
                password.append(chr(random.randint(65,90)))
            elif choice == 3:
                password.append(chr(random.randint(97,122)))
            elif choice == 4:
                password.append(random.choice(["%", "@", "*", "#", "&", "?"]))
    random.shuffle(password)
    password_string = "".join(password)
    return password_string
```

```python
print(password(8))
```
