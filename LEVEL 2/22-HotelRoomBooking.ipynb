## Hotel room booking CLI


```python
import datetime

class Hotel():

    def __init__(self, hotelName, location, rooms, bookedRooms, roomCharge):
        self.hotelName = hotelName
        self.location = location
        self.rooms = rooms
        self.bookedReservations = {room: [] for room in rooms}
        self.roomCharge = roomCharge

    def checkAvailability(self, roomNo, stayFrom: datetime.date, stayUpto: datetime.date):
        if roomNo not in self.rooms:
            print(f"No such Room {roomNo} present in {self.hotelName}!")
            return False

        if stayFrom >= stayUpto:
            print("Error: Stay-From date must be before Stay-Upto date.")
            return False

        reservations = self.bookedReservations[roomNo]

        for existing_start, existing_end in reservations:
            if (stayFrom < existing_end) and (stayUpto > existing_start):
                print(f"Room {roomNo} is booked from {existing_start.strftime('%d-%b')} to {existing_end.strftime('%d-%b')}.")
                print("Booking failed: Requested dates overlap with an existing reservation.")
                return False

        print(f"Room {roomNo} is Open for Booking from {stayFrom.strftime('%d-%b')} to {stayUpto.strftime('%d-%b')}!")
        return True


def bookRoom(self, roomNo, stayFrom: datetime.date, stayUpto: datetime.date):
    available = self.checkAvailability(roomNo, stayFrom, stayUpto)

    if available:
        self.bookedReservations[roomNo].append((stayFrom, stayUpto))
        print(
            f"Success! Room {roomNo} booked for {stayFrom.strftime('%d-%b')} to {stayUpto.strftime('%d-%b')}."
        )
        return True
    else:
        return False
```

```python
rooms = [101, 102, 103, 104, 105, 201, 202, 203, 204, 205, 206, 207, 208, 301, 302, 303, 304, 305, 306, 307, 308, 401, 402, 403, 404, 405, 406, 407, 408]
dasResidency = Hotel("Das Residency", "3C, Naskarpara Lane, Sahidnagar, Dhakuria, Kolkata", rooms, [], 800)
```
