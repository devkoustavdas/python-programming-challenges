```python
import random

class Contacts():
    def __init__(self, name, contactList = None):
        self.name = name
        self.contactList = contactList or {}

    def addContact(self, contactName, category, companyName, position, phoneNo, emailID):
        contactID = (
            chr(random.randint(65, 90))
            + chr(random.randint(65, 90))
            + chr(random.randint(65, 90))
        )
        self.contactList[contactID] = {
            "Name": contactName,
            "Category": category,
            "Company": companyName,
            "Position": position,
            "Phone": phoneNo,
            "Email": emailID
        }
        self.notification("+")
        return contactID, self.contactList[contactID]

    def removeContact(self, contactID):
        self.contactList.pop(contactID)
        self.notification("-")

    def checkDetails(self, contactID):
        c = self.contactList[contactID]
        print(f"{contactID} : {c['Name']} ({c['Category']})")
        print(f"{c['Position']}, {c['Company']}")
        print(f"{c['Phone']} | {c['Email']}")

    def notification(self, target):
        action = ""
        if target == "+":
            action = "Added"
        elif target == "-":
            action = "Removed"
        print(f"Contact {action} successfully!")
```

```python
koustav = Contacts("Koustav Das")
```

```python
koustav.addContact("Michael Jordan", "Friends", "Basketball Association", "Senior Player", 8989676754, "michael.jordan@email.com")
```

```python
koustav.checkDetails("BFX")
```
