## Exhaustive Search

- Given equation: x<sup>2</sup>+54/x
- Use Exhaustive Search method to give the range in which the minima of the given equation lies.

```python
a = 0
b = 5
n = 10
```

```python
def f(x):
    if x == 0:
        x = 10e-12
    return x * x + 54 / x
```

```python
def minimizeExhaustiveSearchMethod(a, b, n):
    h = (b - a) / n
    x = [a, a + h, a + 2 * h]
    fn = [f(x[0]), f(x[1]), f(x[2])]

    for i in range(n - 2):
        print(i)
        if fn[0] > fn[1] and fn[1] < fn[2]:
            return [x[0], x[2]]
        elif x[2] < b:
            x[0] = x[1]
            x[1] = x[2]
            x[2] = x[1] + h
            fn[0] = fn[1]
            fn[1] = fn[2]
            fn[2] = f(x[2])
```

```python
result1 = minimizeExhaustiveSearchMethod(a, b, n)
print(f"Minima at Interval: {result1[0]} - {result1[1]}")
```
