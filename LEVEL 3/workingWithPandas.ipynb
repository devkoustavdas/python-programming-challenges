```python
import pandas as pd
```

```python
df = pd.DataFrame(data_records)
```

```python
df.info()
```

```python
df["Salary"] = df["Salary"].fillna(0)
df["SecondaryIncome"] = df["SecondaryIncome"].fillna(0)
df["TotalIncome"] = df["Salary"] + df["SecondaryIncome"]
```

```python
df.head(5)
```

```python
round(df.describe())
```

```python
df.sort_values(by="TotalIncome", ascending=False, inplace=True)
```

```python
male_records_df = df[df["Gender"] == 'M'].copy()
```

```python
female_records_df = df[df["Gender"] == 'F'].copy()
```

```python
print(f"No. of Male Workers: {male_records_df.shape[0]}")
print(f"No. of Male Workers having Second Job: {male_records_df["SecondJob"].notna().sum()}")
print(f"Mean Income of Male Workers: ${round(male_records_df["TotalIncome"].mean(), 2)}")
print(f"Median Income of Male Workers: ${round(male_records_df["TotalIncome"].median(), 2)}")
print(f"Average Yrs of Experience: {round(male_records_df["YearsExperience"].mean(), 1)}")
```

```python
median_income = df["TotalIncome"].median()
low_income_workers = df[df["TotalIncome"] < median_income]

print(f"Median income: {median_income}")
print("Workers earning below median income:")
print(low_income_workers[["Names", "TotalIncome", "JobRole"]])
```

```python
mean_income = df["TotalIncome"].mean()
low_income_employees = df[df["TotalIncome"] < mean_income]

print(f"Mean income: {mean_income}")
print("Workers earning below mean income:")
print(low_income_employees[["Names", "TotalIncome", "JobRole"]])
```
