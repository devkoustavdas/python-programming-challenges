## Replace Duplicates
- Write a Python function that replaces every group of consecutive duplicate characters in a given string with a single # symbol.
- If multiple groups of duplicates appear consecutively after replacement, they should also be collapsed into a single #.
- Maintain the order of non-duplicate characters.

```python
def replaceDuplicates(text):
    replacedWord, result = "", ""
    i = 0
    while i < len(text):
        if i < len(text) - 1 and text[i] == text[i + 1]:
            while i < len(text) - 1 and text[i] == text[i + 1]:
                i += 1
            replacedWord += "#"
        else:
            replacedWord += text[i]
        i += 1

    i = 0
    while i < len(replacedWord):
        if i < len(replacedWord) - 1 and replacedWord[i] == replacedWord[i + 1]:
            while i < len(replacedWord) - 1 and replacedWord[i] == replacedWord[i + 1]:
                i += 1
            result += "#"
        else:
            result += replacedWord[i]
        i += 1
    return result
```

```python
replaceDuplicates("aaeebeeecihhiaannhkhhhaab")
```
