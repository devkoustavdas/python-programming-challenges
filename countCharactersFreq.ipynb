## Count Character Frequency
- Write a function that takes a string and returns a dictionary with the frequency of each character.
- Sort the characters in decending order

```python
def countCharactersFreq(text):
    text = text.strip().replace(" ", "").upper()
    charFreq = {}
    for char in text:
        if char.isalpha():
            if char not in charFreq:
                charFreq[char] = 0
            charFreq[char] += 1

    charFreq = dict(sorted(charFreq.items(), key=lambda character : character[1], reverse=True))
    for key, value in charFreq.items():
        print(f"{key} appeared {value} times")
```

```python
countCharactersFreq(
    "Kolkata has several interesting facts, including being the former capital of British India and home to the world's oldest polo club. It features the world's busiest cantilever bridge, Howrah Bridge, has Asia's oldest electric tram system, and was the first city in India to have an underground metro."
)
```
