## Days Until Birthday

```python
import datetime
import math

def days_until_birthday(birthday: datetime.date):
    today = datetime.datetime.now()

    age = today.year - birthday.year
    if (today.month, today.day) < (birthday.month, birthday.day):
        age -= 1

    try:
        next_birthday_date = datetime.datetime(today.year, birthday.month, birthday.day, 0, 0, 0)
    except ValueError:
        if birthday.month == 2 and birthday.day == 29:
            next_birthday_date = datetime.datetime(today.year, 2, 28, 0, 0, 0)
        else:
            raise

    if next_birthday_date < today:
        try:
            next_birthday_date = datetime.datetime(today.year + 1, birthday.month, birthday.day, 0, 0, 0)
        except ValueError:
            if birthday.month == 2 and birthday.day == 29:
                next_birthday_date = datetime.datetime(today.year + 1, 2, 28, 0, 0, 0)
            else:
                raise

    time_difference = next_birthday_date - today

    days_left = time_difference.days

    hours_left = math.ceil(time_difference.total_seconds() / 3600)

    print(f"Current Age: {age}")
    print(f"Next Birthday Date: {next_birthday_date.strftime('%d %B, %Y')}")
    print(f"Days Left: {days_left}")
    print(f"Hours Left: {hours_left}")
```

```python
birth_date = datetime.date(1990, 12, 1)
days_until_birthday(birth_date)
```
